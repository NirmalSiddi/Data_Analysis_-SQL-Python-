```python
# read data from the file and handle null values
import pandas as pd
df = pd.read_csv('F:/Portfolio/Sql Projects/sql_python/orders.csv', na_values=['Not Available','unknown']) 
#use na_values to rename not available or unknown or any other to na
df.head(10)
```

```python
# handle null values - check previous step na_values
df['Ship Mode'].unique()
```

```python
# rename column names and lower case them and replace s[ace with string
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.head(10)
```

```python
# derive new column discount, sale price and profit
df['discount'] = df['list_price']*df['discount_percent']*0.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] =  df['sale_price'] - df['cost_price']
df.head(10)
```

```python
df.dtypes
```

```python
#date columns nned to be corrected to date datatype
df['order_date'] = pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
df.dtypes
```

```python
# drop cost proce, list price, discount percent column
df.drop(columns=['list_price','cost_price','discount_percent'], inplace = True)
```
